**Q1.** Please answer the following questions.
1. What is the difference between regression and classification?

    *   Regression is a type of supervised machine learning that predicts continoous numerical values. For example, based on an input, it could predict a house price or temperature
    *   Classification is a type of supervised machine learning that predicts categorical labels. For example, based on an input, it could predict whether someone was likely sick or healthy, or if a home was a 4-bed or a 3-bed.

---

2. What is a confusion table/matrix? What does it help us understand about a model's performance?

    *  A confusion table/matrix is a table that compares a model's predicted class labels to the actual class labels
    *  The matrix is vital in understanding a model's performance because, not only does it show how often the model is wrong, it also shows where the model is most commonly wrong. For example, there could be some classfications for which the model is extremely accurate, and some for which the model is more commonly getting wrong. This allows for specific diagnosis of the problem and easier identification of biases.

---

3. What is Accuracy? Why might it not be entirely sufficient to evaluate a classifer's predictive performance?

    *  Accuracy is the proportion of correct predictions out of all predictions: $$Accuracy = \frac{Number~of~Correct~Predictions}{Number~of~Total~Predictions}$$
    *  Accuracy in general isn't entirely sufficient to evaluate a classifier's predictive performance as it only gives you the overall accuracy and doesn't show where the model is succesful and where it struggles. Additionally, if a classifier guesses everything is one class, and that class has the majority, the accuracy will be high but there will be no predictive power.

---

4. What does the root mean squared error quantify about a particular model?

    *  The root mean squared error quantifies the typical size of prediction errors in a model. Essentially, it is the average error between the predicted value and the real value. Because the value is squared, the large errors are weighted more heavily.

---

5. What are overfitting and underfitting?

    *  Overfitting is when a model learns the training data too closely, including noise and random variation. It performs very well on training data but poorly on new, unseen data. In $k-NN$, this often happens when $k$ is too small.
    *  Underfitting is when a model is too simple to capture the underlying pattern in the data. It performs poorly on both training and test data. In $k-NN$, this often happens when $k$ is too large.

---    

6. Why does splitting the data into training and testing sets, and choosing $k$ by evaluating accuracy or RMSE on the test set, improve model performance?
   
   *  Splitting the data into training and testing sets allows for an evaluation of how well the model generalizes to unseen data. If $k$ is chosen based only on training performance, the model may overfit, meaning it performs very well on the training data but poorly on new observations. By training the model on one subset of the data and evaluating its performance on a separate test set, a more realistic estimate of predictive performance is obtained. Selecting $k$ based on test accuracy or RMSE helps balance overfitting and underfitting, leading to better generalization.

---    

7. With classification, we can report a class label as a prediction or a probability distribution over class labels. Please explain the strengths and weaknesses of each approach.

    * Predicting a class label produces a single category as the output. This approach is simple and easy to interpret, and it is useful when a definite decision must be made. However, it does not convey how confident the model is in its prediction and hides uncertainty.

    * Predicting a probability distribution provides the likelihood of each possible class. This approach communicates model confidence and allows decision-makers to adjust thresholds depending on the costs of different errors. However, probabilities can be more complex to interpret and require proper calibration to be meaningful.

````python
from sklearn.model_selection import train_test_split

X = df[['voltage', 'height', 'soil']]
y = df['mine_type']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.5,
    random_state=100,
    stratify=y
)
````

The dataset was split into 50% training data and 50% test data. Because this is a multi-class classification problem with five mine types, stratified sampling was used to ensure that each class was proportionally represented in both the training and test sets. ##**Question 2 - Part 3**


````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

k_values = range(1, 101)
accuracies = []

for k in k_values:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracies.append(accuracy_score(y_test, y_pred))

plt.figure()
plt.plot(k_values, accuracies)
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Test Accuracy")
plt.title("k-NN Test Accuracy vs. k")
plt.show()

best_k = k_values[accuracies.index(max(accuracies))]
best_accuracy = max(accuracies)

print("Best k:", best_k)
print("Best Accuracy:", best_accuracy)
````

The optimal value of $k$ was determined by evaluating test set accuracy for values of $k$ ranging from 1 to 100. The value that maximized test accuracy was selected, which ended up being a $k$ value of 3 with an accuracy of 0.43.

````python
X = data[['year', 'mileage']]
y = data['price']

X_train_cars, X_test_cars, y_train_cars, y_test_cars = train_test_split(X, y, test_size=0.2, random_state=100)

print(y_train_cars.shape)
print(y_test_cars.shape)
````

Shows that the 2499 rows are successfully split 1999/500 (4:1)

##**Question 3 - Part 4**

````python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

k_values_cars = [3, 10, 25, 50, 100, 300]

mse_list = []

for k in k_values_cars:
    knn_model = KNeighborsRegressor(n_neighbors=k)

    knn_model.fit(X_train_cars, y_train_cars)

    y_pred_cars = knn_model.predict(X_test_cars)

    mse = mean_squared_error(y_test_cars, y_pred_cars)
    mse_list.append(mse)

    plt.figure(figsize=(6,5))
    plt.scatter(y_test_cars, y_pred_cars, alpha=0.5)
    plt.plot([y_test_cars.min(), y_test_cars.max()],
             [y_test_cars.min(), y_test_cars.max()], 'r--')
    plt.xlabel('Actual Price')
    plt.ylabel('Predicted Price')
    plt.title(f'k-NN Regression: k={k}, MSE={mse:.2f}')
    plt.show()

for k, mse in zip(k_values_cars, mse_list):
    print(f"k={k}, MSE={mse:.2f}")
````

The MSE values decrease as $k$ increases up until 100. After this (or potentially before it) the increasing $k$ values begins to increase MSE. This means the optimal $k$ to minimize the MSE lies somewhere between 50 and 300. The MSE values will be plotted vs $k$ in part 5, assuming the best value lies somewhere from 75 - 200. This value would appear as a turning point in the slope, and if this isn't present, the range will be expanded.

##**Question 3 - Part 5**

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

k_range_cars = range(75, 200)
mse_scores = []

for k in k_range_cars:
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train_cars, y_train_cars)
    y_pred_car = knn.predict(X_test_cars)
    mse_scores.append(mean_squared_error(y_test_cars, y_pred_car))

optimal_k = k_range_cars[np.argmin(mse_scores)]
min_mse = min(mse_scores)

plt.figure(figsize=(10, 6))
plt.plot(k_range_cars, mse_scores, marker='o', linestyle='dashed', color='royalblue', markersize=4)
plt.axvline(x=optimal_k, color='red', linestyle='--', label=f'Optimal k = {optimal_k}')
plt.title('MSE vs. k-Value (Finding the Optimal k)')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Mean Squared Error')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"The optimal k value is {optimal_k} with an MSE of {min_mse:,.2f}")
````
